```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("https://raw.githubusercontent.com/ine-rmotr-curriculum/FreeCodeCamp-Pandas-Real-Life-Example/master/data/sales_data.csv", parse_dates=['Date'])
```

```python
### product analysis
### Most Ordered Product
orders_by_product = sales.groupby('Product')['Order_Quantity'].sum().reset_index()
orders_sorted = orders_by_product.sort_values(by="Order_Quantity", ascending=False)
top_orders = orders_sorted.head()
top_orders.plot(kind="bar", x="Product", y="Order_Quantity")
```

```python
### Highest Mark Up
from matplotlib.ticker import FuncFormatter
```

```python
sales["mark_up"] = sales['Unit_Price'] - sales["Unit_Cost"]
sales_grouped = sales.groupby('Product')['mark_up'].sum().reset_index()
sales_grouped = sales_grouped.sort_values(by="mark_up", ascending=False)
mark_up_head = sales_grouped.head()
mark_up_head.plot(kind="bar", x="Product", y="mark_up").yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:.0f}K'))
```

```python
### least revenue generating products
orders_by_product = sales.groupby('Product')['Revenue'].sum().reset_index()
orders_sorted = orders_by_product.sort_values(by="Revenue", ascending=True)
revenue_top = orders_sorted.head()
revenue_top.plot(kind="bar", x="Product", y="Revenue").yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:.0f}K'))
```

```python
### most revenue generating products
orders_by_product = sales.groupby('Product')['Revenue'].sum().reset_index()
orders_sorted = orders_by_product.sort_values(by="Revenue", ascending=True)
top_rev = orders_sorted.tail()
top_rev.plot(kind='bar', x='Product', y="Revenue").yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:.0f}K'))
```
